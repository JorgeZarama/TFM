```python
import requests
import pandas as pd

# Tu API key de Finnhub
api_key = 'cr11209r01qo0i56kq4gcr11209r01qo0i56kq50'

# Definimos las acciones que vamos a analizar
symbols = ['ASML', 'GOOGL', 'LRCX', 'TSM']

# Fechas de inicio y fin
start_date = '2018-08-01'
end_date = '2024-08-01'

# Inicializar un DataFrame para guardar los datos
columns = ['Symbol', 'Date', 'P/E', 'P/B', 'Dividend Yield']
df = pd.DataFrame(columns=columns)

# Función para obtener datos fundamentales históricos
def get_fundamental_data(symbol, api_key, start_date, end_date, df):
    url = f'https://finnhub.io/api/v1/stock/metric?symbol={symbol}&metric=all&token={api_key}'
    r = requests.get(url)
    data = r.json()
    if 'metric' in data:
        dates = pd.date_range(start=start_date, end=end_date, freq='M')
        for date in dates:
            df_row = {
                'Symbol': symbol,
                'Date': date.strftime('%Y-%m-%d'),
                'P/E': data['metric'].get('peInclExtraTTM', None),
                'P/B': data['metric'].get('pbAnnual', None),
                'Dividend Yield': data['metric'].get('dividendYieldIndicatedAnnual', None)
            }
            df = df.append(df_row, ignore_index=True)
    return df

# Obtener y guardar los datos en el DataFrame
for symbol in symbols:
    df = get_fundamental_data(symbol, api_key, start_date, end_date, df)

# Guardar en un archivo CSV
df.to_csv('fundamental_data_finnhub.csv', index=False)

print("Datos guardados en 'fundamental_data_finnhub.csv'")
```
