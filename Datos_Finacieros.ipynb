```python
import yfinance as yf
import pandas as pd

# Definir el símbolo de la acción
symbol = 'ASML'

# Descargar los datos desde el 1 de enero de 2023 hasta el 31 de diciembre de 2023
data = yf.download(symbol, start='2018-08-01', end='2024-08-01', interval='1d')

# Mostrar las primeras filas del dataframe
print(data.head())
```

```python
display(data.head())
```

```python
display(data.describe(include='all'))
```

```python
data.info()
```

```python
data.to_pickle('ASML.pkl')
```
